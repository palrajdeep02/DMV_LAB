```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Create figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Create a circle
circle, = ax.plot([], [], 'bo', markersize=15)

# Initialize function
def init():
    circle.set_data([], [])
    return circle,

# Animation function
def update(frame):
    x = frame
    y = 5 + 2 * np.sin(frame)  # moves in a wave pattern
    circle.set_data(x, y)
    return circle,

# Create animation
ani = FuncAnimation(
    fig,
    update,
    frames=np.linspace(0, 10, 100),
    init_func=init,
    blit=True,
    interval=50
)

plt.show()
```
