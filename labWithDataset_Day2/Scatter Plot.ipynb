```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibility
np.random.seed(42)

# --- Cluster 1 (positive correlation) ---
x1 = np.random.normal(20, 5, 50)
y1 = x1 + np.random.normal(0, 3, 50)

# --- Cluster 2 (another group, shifted) ---
x2 = np.random.normal(50, 5, 50)
y2 = x2 + np.random.normal(0, 3, 50)

# --- Outliers ---
x_out = np.array([5, 90])
y_out = np.array([80, 10])

# Combine all data
x = np.concatenate([x1, x2, x_out])
y = np.concatenate([y1, y2, y_out])

# Plot
plt.figure()
plt.scatter(x, y)

plt.title("Scatter Plot with Clusters, Correlation and Outliers")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```
