```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
file_path = "rainfall in india 1901-2015.csv"
df = pd.read_csv(file_path)

# Step 2: Select only required columns
cols = ['JAN', 'FEB', 'MAR', 'APR']
data = df[cols]

# Step 3: Create boxplot WITHOUT outliers
plt.figure()
data.boxplot(showfliers=False)

# Step 4: Add labels and title
plt.title("Monthly Rainfall Distribution (Without Outliers)")
plt.xlabel("Months")
plt.ylabel("Rainfall")

# Step 5: Display plot
plt.show()
```
